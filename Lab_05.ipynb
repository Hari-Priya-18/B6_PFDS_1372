```python
import pandas as pd
```

```python
products = pd.DataFrame({
    'product_id': [101, 102, 103, 104, 105],
    'product_name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Printer'],
    'category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Electronics'],
    'price': [60000, 800, 1500, 12000, 7000]
})

sales = pd.DataFrame({
    'sale_id': [1, 2, 3, 4, 5, 6],
    'product_id': [101, 102, 103, 104, 101, 105],
    'quantity': [2, 5, 3, 1, 1, 2]
})

products, sales
```

```python
print(products.head())
print(sales.head())
print("Unique product_ids in products:", products['product_id'].nunique())
print("Unique product_ids in sales:", sales['product_id'].nunique())
```

```python
inner_merge = pd.merge(sales, products, on='product_id', how='inner')
inner_merge
```

```python
inner_merge['revenue'] = inner_merge['quantity'] * inner_merge['price']
total_revenue = inner_merge['revenue'].sum()
print("Total Revenue:", total_revenue)
avg_price_simple = inner_merge['price'].mean()
print("Average Price:", avg_price_simple)
```

```python
product_sales = inner_merge.groupby('product_name', as_index=False).agg(
    total_quantity=('quantity', 'sum'),
    total_revenue=('revenue', 'sum')
).sort_values(['total_revenue', 'total_quantity'], ascending=False)

product_sales
```

```python
category_stats = inner_merge.groupby('category', as_index=False).agg(
    total_quantity=('quantity', 'sum'),
    total_revenue=('revenue', 'sum')
).sort_values('total_revenue', ascending=False)

category_stats
```

```python
best_category_by_revenue = category_stats.loc[category_stats['total_revenue'].idxmax(), 'category']
best_category_by_revenue
```
